```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
```

```python
X = torch.tensor([
    [1., 2.],
    [3., 4.],
    [5., 6.],
    [7., 8.],
])

y = torch.tensor([
    [3.],
    [7.],
    [11.],
    [15.]
])

```

```python
class SimpleNet(nn.Module):
    def __init__(self):
        super(SimpleNet, self).__init__()
        self.hidden1 = nn.Linear(in_features=2, out_features=5)
        self.hidden2 = nn.Linear(in_features=5, out_features=10)
        self.hidden3 = nn.Linear(in_features=10, out_features=5)
        self.output = nn.Linear(in_features=5, out_features=1)

    def forward(self, x):
        x = F.relu(self.hidden1(x))
        x = F.relu(self.hidden2(x))
        x = F.relu(self.hidden3(x))
        x = self.output(x)
        return x
```

```python
model = SimpleNet()
lr = 0.0001
optimizer = optim.Adam(model.parameters(), lr=lr)
criterion = nn.MSELoss()


#Training
epochs = 30000
for epoch in range(epochs):

    # Model predict output
    y_pred = model(X)

    # Calculate loss
    loss = criterion(y_pred, y)

    if epoch % 1000 == 0:
        print(f'epoch: {epoch}, loss: {loss}')

    # Reset all gradient
    optimizer.zero_grad()

     # Back propagation to re-calculate parameter
    loss.backward()

    # Calculate gradient
    optimizer.step()

```

```python
X_test = torch.tensor([[10., 2.]])
y_test = torch.tensor([[12.]])

y_pred = model(X_test)
loss = criterion(y_pred, y_test)
print(f"y_pred: {y_pred}")
print(f'loss: {loss}')
```
